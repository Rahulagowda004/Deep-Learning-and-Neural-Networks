```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
ds = pd.DataFrame(data = cancer.data, columns = cancer.feature_names)
```

```python
ds
```

```python
ds.info()
```

```python
ds.describe()
```

```python
cancer.target
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit_transform(ds)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
pca.fit(ds)
```

```python
df = pca.transform(ds)
```

```python
X = df
y = cancer.target
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20, random_state = 0)
```

```python
import keras
from keras import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,ReLU
from keras.layers import Dropout
```

```python
model = Sequential()
```

```python
#adding input layer and first hidden layer
model.add(Dense(9, kernel_initializer = 'he_uniform',activation='relu',input_dim = 10))

#adding second layer
model.add(Dense(9,  kernel_initializer = 'he_uniform', activation = 'relu'))

#adding third layer
model.add(Dense(9, kernel_initializer = 'he_uniform', activation = 'relu'))

#adding output layer
model.add(Dense(1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))

#compiling ANN
model.compile(optimizer = 'Adamax', loss = 'binary_crossentropy' ,  metrics = ['accuracy'])

model_history = model.fit(X_train,y_train, validation_split =  0.33, batch_size = 10, epochs = 100)
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred>0.5)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print(accuracy*100)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```
